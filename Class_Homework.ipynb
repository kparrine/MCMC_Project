# Chi Squared
Equation for it:
$$\chi^{2} = \sum{\frac{(O_{i}-E_{i})^{2}}{\sigma_{i}^{2}}}$$
where $O_{i}$ is the observed value, $E_{i}$ is the predicted model value, and $\sigma_{i}$ is the uncertainty (standard deviation) of the measurement.
